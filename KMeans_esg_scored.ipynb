# machine learning KMeans : analyse exploratoire
- modèle NON SUPERVISE
- quels types d'entreprises existent ils dans mon dataset ? => détection automatique des données => création de labels

```python
# import StandardScaler pour standardiser les features

from sklearn.preprocessing import StandardScaler
```

```python
# définition des variables explicatives

X = df[["Environmental_Score", "Social_Score", "Governance_Score", "ESG_Global_Score"]]
X
```

- mise à l'échelle de X

```python
scaler = StandardScaler().fit(X)
```

```python
X_scaled = scaler.transform(X)
```

```python
X_scaled
```

- instanciation du modèle KMeans

```python
modelKM = KMeans(n_clusters=3, random_state=42)
```

- prédiction et création colonne depuis X mis à l'échelle

```python
df['cluster_ID'] = modelKM.fit_predict(X_scaled)
```

```python
df
```

- analyse des centroides : centroide = la moyenne de toutes les observations du cluster, mais dans l'espace standardisé.

```python
modelKM.cluster_centers_
```

```python
centroid = pd.DataFrame(modelKM.cluster_centers_, columns=X.columns)
print(centroid)
```

```python
# clusters HIGH = 2
# cluster MEDIUM = 1
# cluster LOW = 0
```

- vérification des moyennes par clusters avec un groupby

```python
df.groupby("cluster_ID")[['Environmental_Score', "Social_Score", "Governance_Score", "ESG_Global_Score"]].mean()

# ESG_Global_Score plus haut = cluster HIGH
# ESG_Global_Score moyen = cluster MEDIUM
# ESG_Global_Score plus bas = cluster LOW
```

- mapping des clusters pour création de la colonne dans le df

```python
mapping = {2 : "HIGH", 1 : "MEDIUM", 0 : "LOW"}
df["ESG_Cluster_Label"] = df["cluster_ID"].map(mapping)
```

```python
df
```

```python
df[["ESG_Global_Score", "ESG_Cluster_Label"]].sort_values(by="ESG_Global_Score", ascending=False)
```

- “KMeans segmente selon la forme complète du profil ESG (E/S/G/Global),
pas selon le seul score global.
On observe donc des zones de recouvrement :
certaines entreprises proches de la frontière entre deux clusters ont un score global similaire mais un profil ESG différent.”
